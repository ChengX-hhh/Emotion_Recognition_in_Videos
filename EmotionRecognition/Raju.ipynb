```python
import os
import pandas as pd
import glob
from PIL import Image, ImageDraw, ImageFont
import emotion_detect
```

```python
root_path = '../Video/Raju/'
all_paths = os.listdir(root_path)
all_files = [file for file in all_paths if file.startswith('Raju') ]
all_files
```

```python
for file in all_files:
    path = root_path + file + '/'
    images_files = os.listdir(path)
    file_begin = images_files[0]
    # change order
    images_files.remove(file_begin) 
    images_files.append(file_begin)
    images_selected = [os.path.join(path,file) for file in images_files]
    
    count = 1
    results_df = pd.DataFrame(columns=['name','emotion','frame'])
    for i in images_selected:
        print(i)
        # 1.import image
        img = Image.open(i)
        # 2.face identity
        img_identity = emotion_detect.detect_frame(img)
        save_path = 'Plot_results/Raju/'+file+'/'
        if not os.path.exists(save_path):
            os.makedirs(save_path)
        img_identity[0].save(save_path+str(count)+'.png')

        temp_df = img_identity[1]
        temp_df['frame'] = count
        results_df = pd.concat([results_df,temp_df],axis=0)
        count = count + 1
    save_data_path = 'Data_results/Raju/'
    if not os.path.exists(save_data_path):
        os.makedirs(save_data_path)
    results_df.to_csv(save_data_path+file+'.csv')
```
